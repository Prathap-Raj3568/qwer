```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
vgsales = pd.read_csv("vgsales.csv") 
vgsales.head(10)
```

```python
vgsales.isnull().values.any()
```

```python
print("Database dimension     :",vgsales.shape)
print("Database size          :",vgsales.size)
vgsales.info()
vgsales.describe()
```

```python
games =vgsales['Name'].unique()
publishers =vgsales ['Publisher'].unique()
platforms =vgsales ['Platform'].unique()
game_type = vgsales['Genre'].unique()

print("Number of Games: ",len(games))
print("Publishers: ",len(publishers))
print("Platforms: ",len(platforms))
print("Game Types: ",len(game_type))
```

```python
fig,ax = plt.subplots(figsize=(5,5))
vgsales['Genre'].value_counts(sort=False).plot(kind='barh',color = 'blue')
plt.title('Genre Distribution',fontsize=10)
plt.xlabel('Genre',fontsize=10)
plt.ylabel('Number of sales',fontsize=10)
```

```python
fig,ax = plt.subplots(figsize=(10,10))
vgsales['Year'].value_counts(sort=True).plot(kind='bar',color = "red")
plt.title('Year Distribution',fontsize=15)
plt.xlabel('Year',fontsize=15)
plt.ylabel('Number of sales',fontsize=15)
```

```python
fig,ax = plt.subplots(figsize=(20,20))
vgsales['Platform'].value_counts(sort=True).plot(kind='pie')
plt.title('Platform Distribution',fontsize=20)
plt.xlabel('Platform',fontsize=20)
plt.ylabel('Number of sales',fontsize=20)
```
